```python
import tiktoken
```

```python
enc = tiktoken.encoding_for_model('gpt-4')
```

```python
tokens = enc.encode('Generative AI is transforming the world!')
```

```python
print(f'Tokens: {tokens}')
print(f'Token count: {len(tokens)}')
print(f'Decoded: {enc.decode(tokens)}')
```

EXPLAINING THE ABOVE CODE-
What is happening?
LLMs don’t understand text directly
They convert text → tokens (numbers)
Tokenization is how GPT “sees” text
encode() → text to numbers
decode() → numbers back to text

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
encoded = tokenizer('Hello, I am learning about LLMs!', return_tensors='pt')
print('Input IDs:', encoded['input_ids'])
print('Tokens:', tokenizer.convert_ids_to_tokens(encoded['input_ids'][0]))
```

the code below is Text generation (how GPT-2 writes text)



```python
from transformers import pipeline
import torch
```

```python
generator = pipeline(
 'text-generation',
 model='gpt2',
 device=-1
)

#This is a ready-made LLM text generator
```

```python
prompt = 'Generative AI will transform education because' #And it continues writing after the prompt
outputs = generator(
 prompt,
 max_new_tokens=100, #generate up to 100 new tokens
 num_return_sequences=3,
 temperature=0.8,  # Randomess / creativity
 do_sample=True,
 top_p=0.95,#choose from smallest set of words that make up 95% probability
 top_k=50#choose from top 50 words
)

```

#It gives 3 completions:

Option 1
Option 2
Option 3

 This is probabilistic text generation


```python
for i, output in enumerate(outputs):
 print(f'\n--- Option {i+1} ---')
 print(output['generated_text'])
```

#the below code is of IBM Granite LLM (real instruction model)

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
model_id = 'ibm-granite/granite-3.1-2b-instruct' #This is a real instruction-tuned LLM (like ChatGPT-style model).
tokenizer = AutoTokenizer.from_pretrained(model_id) #loading tokenizer + model
model = AutoModelForCausalLM.from_pretrained(
 model_id,
 torch_dtype=torch.float16,
 device_map='auto'
)

#What happens?
#tokenizer → text → tokens
#model → predicts next token repeatedly
```

```python
messages = [
 {'role': 'user',
 'content': 'Explain what a transformer is in 3 bullet points for a student.'}
]
#Chat format input
```

```python
input_ids = tokenizer.apply_chat_template(messages, return_tensors='pt')['input_ids'].to(model.device)
#Convert chat → tokens
output = model.generate(
 input_ids,
 max_new_tokens=256,
 temperature=0.7,
 do_sample=True
)
#it predicts:token1 → token2 → token3 → ... until max tokens reached.

```

```python
response = tokenizer.decode(output[0][input_ids.shape[1]:], skip_special_tokens=True)
print(response)

#decoding--Removes input prompt and shows only answer.
```
