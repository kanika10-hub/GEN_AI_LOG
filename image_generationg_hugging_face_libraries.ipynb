```python
from huggingface_hub import HfApi , list_models
```

```python
api=HfApi()
```

```python
models = list(
    list_models(
        filter='text-generation',
        sort='downloads',
        limit=5
    )
)
```

```python
print('Top 5 Text Generation Models:')
for m in models:
    print(f' {m.modelId} — {m.downloads:,} downloads')
```

```python
model_info = api.model_info('ibm-granite/granite-3.1-2b-instruct')
print('\nModel Tags:', model_info.tags)
```

```python
from huggingface_hub import InferenceClient
```

```python
client = InferenceClient(token='HUGGING_FACE_TOKEN')
```

```python
print('\n--- Image Generation (via InferenceClient) ---')
try:
  image = client.text_to_image(
      model="stabilityai/stable-diffusion-xl-base-1.0",
      prompt='An apple on a cherry blossom tree and a kid sitting under a tree'
  )

  if image:
    image.save('generated_image_client1.jpg')
    print('Image saved as generated_image_client1.jpg!')
  else:
    print('Image generation returned no image.')
except Exception as e:
  print(f"Error during image generation with InferenceClient: {e}")
```
