```python
from sklearn.datasets import fetch_california_housing
housing=fetch_california_housing(as_frame=True)
real_data=housing.frame
print(real_data.shape,real_data.describe())
```

```python
from ctgan import CTGAN
model=CTGAN(epochs=10,verbose=True)
model.fit(real_data,[])
synthetic_data=model.sample(500)
print(synthetic_data.describe())
```

```python
from table_evaluator import TableEvaluator
evaluator=TableEvaluator(real_data.sample(500),synthetic_data)
evaluator.visual_evaluation()
evaluator.evaluate(target_col='MedHouseVal')
```
